***Timeseries Analysis***

```python
def timeseries_analysis_plot(data, timestamp_col, ghi_col, dni_col, dhi_col, tamb_col):
    """
    Perform time series analysis and plot GHI, DNI, DHI, and Tamb over time.

    Args:
    data (DataFrame): DataFrame containing the time series data.
    timestamp_col (str): Name of the column containing timestamps.
    ghi_col (str): Name of the column containing GHI data.
    dni_col (str): Name of the column containing DNI data.
    dhi_col (str): Name of the column containing DHI data.
    tamb_col (str): Name of the column containing Tamb data.

    Returns:
    None (plots the analysis).
    """
    # Convert timestamp column to datetime format
    data[timestamp_col] = pd.to_datetime(data[timestamp_col])

    # Set timestamp column as index
    data.set_index(timestamp_col, inplace=True)

    # Plotting the variables over time
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.plot(data.index, data[ghi_col], color='blue')
    plt.title('Global Horizontal Irradiance (GHI)')
    plt.xlabel('Time')
    plt.ylabel('GHI')

    plt.subplot(2, 2, 2)
    plt.plot(data.index, data[dni_col], color='green')
    plt.title('Direct Normal Irradiance (DNI)')
    plt.xlabel('Time')
    plt.ylabel('DNI')

    plt.subplot(2, 2, 3)
    plt.plot(data.index, data[dhi_col], color='red')
    plt.title('Diffuse Horizontal Irradiance (DHI)')
    plt.xlabel('Time')
    plt.ylabel('DHI')

    plt.subplot(2, 2, 4)
    plt.plot(data.index, data[tamb_col], color='orange')
    plt.title('Ambient Temperature (Tamb)')
    plt.xlabel('Time')
    plt.ylabel('Tamb')

    plt.tight_layout()
    plt.show()
```

```python
timeseries_analysis_plot(df1, 'Timestamp', 'GHI', 'DNI', 'DHI', 'Tamb')
```

```python
correlation_matrix = df1[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_matrix_wind =df1[['WS', 'WSgust', 'WSstdev', 'WD', 'WDstdev']].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_wind, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix (Wind)')
plt.show()

# Plotting wind speed over time
plt.figure(figsize=(12, 8))
plt.plot(df1.index, df1['WS'], color='blue', label='Wind Speed (WS)')
plt.plot(df1.index, df1['WSgust'], color='green', label='Wind Gust Speed (WSgust)')
plt.plot(df1.index, df1['WSstdev'], color='red', label='Wind Speed Standard Deviation (WSstdev)')
plt.title('Wind Speed')
plt.xlabel('Time')
plt.ylabel('Wind Speed (m/s)')
plt.legend()
plt.show()

# Plotting wind direction over time
plt.figure(figsize=(12, 6))
plt.plot(df1.index, df1['WD'], color='blue', label='Wind Direction (WD)')
plt.plot(df1.index, df1['WDstdev'], color='red', label='Wind Direction Standard Deviation (WDstdev)')
plt.title('Wind Direction')
plt.xlabel('Time')
plt.ylabel('Wind Direction (degrees)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_temp, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix (Temperature)')
plt.show()

# Plotting module temperatures and ambient temperature over time
plt.figure(figsize=(12, 8))
plt.plot(df1.index, df1['TModA'], color='blue', label='Module Temperature A (TModA)')
plt.plot(df1.index, df1['TModB'], color='green', label='Module Temperature B (TModB)')
plt.plot(df1.index, df1['Tamb'], color='red', label='Ambient Temperature (Tamb)')
plt.title('Temperature Comparison')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```
